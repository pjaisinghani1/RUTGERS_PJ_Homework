```python
#MatPlotLib Notebook
%matplotlib notebook

# Dependencies and Setup
import os
import csv
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas Data Frames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])

school_data_complete.head()
```

```python
# Replace all NaN values with 0 
school_data_complete = school_data_complete.fillna(0)

#Calculate the number of unique schools in the Dataframe 
total_schools = school_data_complete['school_name'].nunique()

total_schools
```

```python
#Calculate the total number of students
total_students = school_data_complete["student_name"].count()
total_students
```

```python
#Calculate the total budget by first getting all the unique budgets of each school district
unique_budgets = list(school_data_complete["budget"].agg(['unique']))

unique_budgets
```

```python
#Convert array to list
budget_list = np.array(unique_budgets).tolist()
```

```python
#Calculate total budget 
total_budget= np.array(budget_list).sum()
total_budget
```

```python
#Calculate  the average math score 

average_math_score= school_data_complete["math_score"].mean()
average_math_score
```

```python
#Calculate  the average reading score 

average_reading_score= school_data_complete["reading_score"].mean()
average_reading_score

```

```python
#Calculate the % passing math 
 # of Passing Students 
    
#Create Bins to group scores 
bins = [0, 65, 69, 79, 89, 100]

# Create the names for the four bins
group_names = ["F", "D", "C", "B", "A"]

#Create a new column with summaries of grades 
school_data_complete["Math Score Summary"] = pd.cut(school_data_complete["math_score"], bins, labels=group_names)

school_data_complete["Reading Score Summary"] = pd.cut(school_data_complete["reading_score"], bins, labels=group_names)

school_data_complete
```

```python
school_data_complete[school_data_complete["Math Score Summary"]!="F"].count()
```

```python
#Count the values in summary columns of students with grades A,B,C or D 

math_passing= school_data_complete[school_data_complete["Math Score Summary"]!="F"].count()
Math_Passing_Percent = (math_passing)/(total_students) * 100
Math_Passing_Percent
```

```python
reading_passing= school_data_complete[school_data_complete["Reading Score Summary"]!="F"].count()

Read_Passing_Percent = (reading_passing)/(total_students) * 100

Read_Passing_Percent
```

```python
#Calculate the overall passing rate (overall average score), i.e. (avg. math score + avg. reading score)/2
print(average_math_score)
print(average_reading_score)

calculation = int(average_math_score) + int(average_reading_score)

overall_average = (calculation)/2
overall_average
```

```python
#Place all data in a summary table
summary_table = pd.DataFrame({"Total Schools": [total_schools],
                              "Total Students": ["{:,}".format(total_students)],
                              "Total Budget": ["${:,.2f}".format(total_budget)],
                              "Average Math Score": ["{:,.2f}".format(average_math_score)],
                              "Average Reading Score": ["{:,.2f}".format(average_reading_score)],
                              "% Passing Math": ["{:,.1f}%".format(Math_Passing_Percent[1])],
                              "% Passing Reading": ["{:,.1f}%".format(Read_Passing_Percent[1])],
                              "% Overall Passing Rate": ["{:,.1f}%".format(overall_average)]
                             })
summary_table
```

```python
#School Names
groupdby_schools = school_data_complete.groupby(["school_name"])
groupdby_schools.count()
```

```python
#Creating a new dataset with math and reading scores to get passing scores and then grouping by school name

school_math_score = school_data_complete.loc[school_data_complete['math_score'] >= 70]
math_score_by_school = school_math_score.groupby('school_name').agg({'math_score':'count'})

school_reading_score = school_data_complete.loc[school_data_complete['reading_score']>=70]
reading_score_by_school = school_reading_score.groupby('school_name').agg({'reading_score':'count'})
students_size = school_data_complete.groupby('school_name').agg({'size':'mean'})

passing_by_school = pd.merge(pd.merge(math_score_by_school, reading_score_by_school, on='school_name'), students_size, on='school_name')

passing_by_school['Math Passing'] = (passing_by_school['math_score'] / passing_by_school['size']) * 100
passing_by_school['Reading Passing'] = (passing_by_school['reading_score'] / passing_by_school['size']) * 100
passing_by_school['Overall Passing'] = (passing_by_school['Math Passing'] + passing_by_school['Reading Passing']) / 2

passing_by_school = passing_by_school[['size', 'Math Passing', 'Reading Passing', 'Overall Passing']]
passing_by_school.head()
```

```python
school_summary = pd.DataFrame()
school_summary["School Type"] = groupdby_schools["type"].first()
school_summary["Total Students"] = groupdby_schools["size"].mean().unique()
school_summary["Total budget"] = groupdby_schools["budget"].mean().unique()
school_summary["Budget/Student"] = groupdby_schools["budget"].mean().unique() / groupdby_schools["size"].mean().unique()
school_summary["Avg Math Score"] = groupdby_schools["math_score"].mean().unique().round()
school_summary["Avg Reading Score"] = groupdby_schools["reading_score"].mean().unique().round()
school_summary["Math Passing %"] = passing_by_school["Math Passing"].round()
school_summary["Reading Passing %"] = passing_by_school["Reading Passing"].round()
school_summary["Overall Passing %"] = passing_by_school["Overall Passing"].round()
school_summary
```
